```python
import pygame
import math
import sys

# Initialize Pygame
pygame.init()

# Screen dimensions
WIDTH, HEIGHT = 800, 400
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Pool Table Simulation")

# Colors
GREEN = (34, 139, 34)
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

# Ball properties
ball_radius = 10
ball_x, ball_y = WIDTH // 2, HEIGHT // 2
ball_velocity = [0, 0]  # [x_velocity, y_velocity]
friction = 0.98  # Friction coefficient

# Game clock
clock = pygame.time.Clock()

# Functions
def draw_table():
    screen.fill(GREEN)
    pygame.draw.rect(screen, BLACK, (50, 50, WIDTH - 100, HEIGHT - 100), 5)  # Table boundary

def draw_ball(x, y):
    pygame.draw.circle(screen, WHITE, (int(x), int(y)), ball_radius)

def move_ball():
    global ball_x, ball_y, ball_velocity
    ball_x += ball_velocity[0]
    ball_y += ball_velocity[1]
    
    # Apply friction
    ball_velocity[0] *= friction
    ball_velocity[1] *= friction

    # Stop the ball if it moves too slowly
    if abs(ball_velocity[0]) < 0.1:
        ball_velocity[0] = 0
    if abs(ball_velocity[1]) < 0.1:
        ball_velocity[1] = 0

    # Collision with walls
    if ball_x - ball_radius <= 50 or ball_x + ball_radius >= WIDTH - 50:
        ball_velocity[0] = -ball_velocity[0]
    if ball_y - ball_radius <= 50 or ball_y + ball_radius >= HEIGHT - 50:
        ball_velocity[1] = -ball_velocity[1]

def hit_ball(angle, power):
    global ball_velocity
    ball_velocity[0] = power * math.cos(angle)
    ball_velocity[1] = -power * math.sin(angle)

# Main loop
running = True
while running:
    screen.fill(GREEN)
    draw_table()
    draw_ball(ball_x, ball_y)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        # Mouse input to hit the ball
        if event.type == pygame.MOUSEBUTTONDOWN:
            mouse_x, mouse_y = pygame.mouse.get_pos()
            dx = mouse_x - ball_x
            dy = mouse_y - ball_y
            angle = math.atan2(-dy, dx)  # Angle to hit
            power = min(math.sqrt(dx**2 + dy**2) / 10, 10)  # Power based on distance (clamped)
            hit_ball(angle, power)

    move_ball()
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
sys.exit()
```

```python
import pygame
import math
import threading

def run_pool_table_simulation():
    # Initialize Pygame
    pygame.init()

    # Screen dimensions
    WIDTH, HEIGHT = 800, 400
    screen = pygame.display.set_mode((WIDTH, HEIGHT))
    pygame.display.set_caption("Pool Table Simulation")

    # Colors
    GREEN = (34, 139, 34)
    WHITE = (255, 255, 255)
    BLACK = (0, 0, 0)
    RED = (255, 0, 0)

    # Ball properties
    ball_radius = 10
    ball_x, ball_y = WIDTH // 2, HEIGHT // 2
    ball_velocity = [0, 0]  # [x_velocity, y_velocity]
    friction = 0.98  # Friction coefficient

    # Button properties
    button_font = pygame.font.SysFont(None, 40)
    button_text = button_font.render('Quit', True, WHITE)
    button_rect = pygame.Rect(WIDTH - 100, HEIGHT - 50, 80, 40)

    # Game clock
    clock = pygame.time.Clock()

    # Functions
    def draw_table():
        screen.fill(GREEN)
        pygame.draw.rect(screen, BLACK, (50, 50, WIDTH - 100, HEIGHT - 100), 5)  # Table boundary

    def draw_ball(x, y):
        pygame.draw.circle(screen, WHITE, (int(x), int(y)), ball_radius)

    def draw_button():
        pygame.draw.rect(screen, RED, button_rect)
        screen.blit(button_text, (button_rect.x + 10, button_rect.y + 5))

    def move_ball():
        nonlocal ball_x, ball_y, ball_velocity
        ball_x += ball_velocity[0]
        ball_y += ball_velocity[1]

        # Apply friction
        ball_velocity[0] *= friction
        ball_velocity[1] *= friction

        # Stop the ball if it moves too slowly
        if abs(ball_velocity[0]) < 0.1:
            ball_velocity[0] = 0
        if abs(ball_velocity[1]) < 0.1:
            ball_velocity[1] = 0

        # Collision with walls
        if ball_x - ball_radius <= 50 or ball_x + ball_radius >= WIDTH - 50:
            ball_velocity[0] = -ball_velocity[0]
        if ball_y - ball_radius <= 50 or ball_y + ball_radius >= HEIGHT - 50:
            ball_velocity[1] = -ball_velocity[1]

    def hit_ball(angle, power):
        nonlocal ball_velocity
        ball_velocity[0] = power * math.cos(angle)
        ball_velocity[1] = -power * math.sin(angle)

    # Main loop
    running = True
    try:
        while running:
            screen.fill(GREEN)
            draw_table()
            draw_ball(ball_x, ball_y)
            draw_button()

            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    running = False

                # Mouse input to hit the ball
                if event.type == pygame.MOUSEBUTTONDOWN:
                    mouse_x, mouse_y = pygame.mouse.get_pos()
                    if button_rect.collidepoint(mouse_x, mouse_y):
                        running = False
                    else:
                        dx = mouse_x - ball_x
                        dy = mouse_y - ball_y
                        angle = math.atan2(-dy, dx)  # Angle to hit
                        power = min(math.sqrt(dx**2 + dy**2) / 10, 10)  # Power based on distance (clamped)
                        hit_ball(angle, power)

            move_ball()
            pygame.display.flip()
            clock.tick(60)
    finally:
        pygame.display.quit()
        pygame.quit()

# Run the simulation in a separate thread
simulation_thread = threading.Thread(target=run_pool_table_simulation)
simulation_thread.start()
```
