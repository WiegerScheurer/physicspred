```python
import pandas as pd
import numpy as np
import os
from collections import defaultdict
from datetime import timedelta, datetime

# Load data
dw_path = "/Users/wiegerscheurer/Library/CloudStorage/OneDrive-RadboudUniversiteit/Donders/Wonders/"
dw_files = os.listdir(dw_path)
team_filename = [f for f in dw_files if "team" in f][0]
planning_filename = [f for f in dw_files if "planning" in f][0]

team_raw_df = pd.read_excel(dw_path + team_filename)
team_df = team_raw_df.loc[0:df.index[df["Name"].isnull()][0]-1]
team_df["Name"] = list(get_first_names(team_df))
# Sort the ddf based on the first name
team_df = team_df.sort_values(by="Name")

planning_df = pd.read_excel(dw_path + planning_filename)

# Generate dates
def generate_dates(start_date, end_date):
    start_date = datetime.strptime(start_date, "%Y-%m-%d")
    end_date = datetime.strptime(end_date, "%Y-%m-%d")
    num_weeks = (end_date - start_date).days // 7
    return [start_date + timedelta(weeks=i) for i in range(num_weeks + 1)]

# Calculate flexibility score (lower = less flexible)
def compute_flexibility(df, role_columns):
    df["Flexibility_Score"] = df[role_columns].apply(lambda row: sum(row == "yes"), axis=1)
    return df.sort_values(by="Flexibility_Score")

# Get all potential people for each role and compute flexibility for all groups
roles = ["Buddy", "Editor Dutch", "Editor English", "Translates English>Dutch", "Translates Dutch > English"]
authors = team_df[team_df["Blogs.1"] == "yes"].copy()
authors["Name"] = list(get_first_names(authors))
authors = compute_flexibility(authors, roles)

dutch_editors = team_df[team_df["Editor Dutch"] == "yes"].copy()
dutch_editors = compute_flexibility(dutch_editors, roles)

english_editors = team_df[team_df["Editor English"] == "yes"].copy()
english_editors = compute_flexibility(english_editors, roles)

translators_en_nl = team_df[team_df["Translates English>Dutch"] == "yes"].copy()
translators_en_nl = compute_flexibility(translators_en_nl, roles)

translators_nl_en = team_df[team_df["Translates Dutch > English"] == "yes"].copy()
translators_nl_en = compute_flexibility(translators_nl_en, roles)

buddies = team_df[team_df["Buddy"] == "yes"].copy()
buddies = compute_flexibility(buddies, roles)

# Function to get the person with fewest assignments in a given role
def get_next_person(eligible_people, workload, role):
    if eligible_people.empty:
        return None
        
    # Sort eligible people by workload
    eligible_people_sorted = eligible_people.copy()
    
    # Add current workload counts to the dataframe
    eligible_people_sorted["Current_Workload"] = eligible_people_sorted["Name"].apply(
        lambda name: workload.get(name, {}).get(role, 0)
    )
    eligible_people_sorted["Total_Workload"] = eligible_people_sorted["Name"].apply(
        lambda name: sum(workload.get(name, {}).values())
    )
    
    # Sort by current role workload, then total workload
    eligible_people_sorted = eligible_people_sorted.sort_values(
        by=["Current_Workload", "Total_Workload"]
    )
    
    # Return the name of the person with the lowest workload
    return eligible_people_sorted.iloc[0]["Name"]

def assign_roles(start_date, end_date):
    dates = generate_dates(start_date, end_date)
    
    # Initialize workload tracking
    workload = defaultdict(lambda: defaultdict(int))
    
    # Initialize assignments dictionary
    assignments = defaultdict(list)
    
    # For each date, assign all roles
    for date in dates:
        # Get author for this date - prioritize those with fewer assignments
        author = get_next_person(authors, workload, "Author")
        
        if author is None:
            print(f"No available author for date {date}")
            continue
            
        # Update workload for this author
        workload[author]["Author"] += 1
        
        # Determine language based on author's capabilities
        is_dutch = team_df[team_df["Name"] == author]["Editor Dutch"].values[0] == "yes"
        
        # Get eligible pools based on language
        buddy_pool = buddies.copy()
        editor_pool = dutch_editors.copy() if is_dutch else english_editors.copy()
        translator_pool = translators_nl_en.copy() if is_dutch else translators_en_nl.copy()
        translation_editor_pool = english_editors.copy() if is_dutch else dutch_editors.copy()
        
        # Exclude the author from other roles for this date
        buddy_pool = buddy_pool[buddy_pool["Name"] != author]
        editor_pool = editor_pool[editor_pool["Name"] != author]
        translator_pool = translator_pool[translator_pool["Name"] != author]
        translation_editor_pool = translation_editor_pool[translation_editor_pool["Name"] != author]
        
        # Get people for each role based on workload
        buddy = get_next_person(buddy_pool, workload, "Buddy")
        
        # Now exclude buddy from other roles
        editor_pool = editor_pool[editor_pool["Name"] != buddy]
        translator_pool = translator_pool[translator_pool["Name"] != buddy]
        translation_editor_pool = translation_editor_pool[translation_editor_pool["Name"] != buddy]
        
        editor = get_next_person(editor_pool, workload, "Editor")
        
        # Exclude editor from other roles
        translator_pool = translator_pool[translator_pool["Name"] != editor]
        translation_editor_pool = translation_editor_pool[translation_editor_pool["Name"] != editor]
        
        translator = get_next_person(translator_pool, workload, "Translator")
        
        # Exclude translator from translation editor role
        translation_editor_pool = translation_editor_pool[translation_editor_pool["Name"] != translator]
        
        translation_editor = get_next_person(translation_editor_pool, workload, "Translation Editor")
        
        # Update workloads
        workload[buddy]["Buddy"] += 1
        workload[editor]["Editor"] += 1
        workload[translator]["Translator"] += 1
        workload[translation_editor]["Translation Editor"] += 1
        
        # Calculate dates
        submit_to_buddy_date = date.strftime("%m/%d/%Y")
        submit_to_editor_translator_date = (date + timedelta(days=7)).strftime("%m/%d/%Y")
        return_to_author_date = (date + timedelta(days=19)).strftime("%m/%d/%Y")
        publish_date = (date + timedelta(days=24)).strftime("%m/%d/%Y")
        
        # Add to assignments
        assignments["Author"].append(author)
        assignments["Submit to Buddy"].append(submit_to_buddy_date)
        assignments["Buddy"].append(buddy)
        assignments["Submit to Editor and Translator"].append(submit_to_editor_translator_date)
        assignments["Editor"].append(editor)
        assignments["Translator"].append(translator)
        assignments["Translation Editor"].append(translation_editor)
        assignments["Return to Author"].append(return_to_author_date)
        assignments["Publish"].append(publish_date)
    
    return assignments, workload

# Run the assignment algorithm
assignments, workload = assign_roles("2025-01-03", "2025-12-04")

# Create and save the assignments DataFrame
assignments_df = pd.DataFrame(assignments)
save_path = dw_path + "output/"
os.makedirs(save_path, exist_ok=True)
assignments_df.to_excel(f"{save_path}Donders_Wonders_Planning_Output.xlsx", index=False)

# Display workload distribution
def analyze_workload(workload):
    # Convert the nested defaultdict to DataFrame for analysis
    workload_rows = []
    for person, roles in workload.items():
        row = {"Name": person}
        total = 0
        for role, count in roles.items():
            row[role] = count
            total += count
        row["Total"] = total
        workload_rows.append(row)
    
    workload_df = pd.DataFrame(workload_rows)
    
    # Fill NaN with 0
    for col in workload_df.columns:
        if col != "Name":
            workload_df[col] = workload_df[col].fillna(0).astype(int)
    
    # Sort by total workload
    workload_df = workload_df.sort_values(by="Total")
    
    print("\n--- Workload Analysis ---")
    print(workload_df)
    print("\nWorkload Statistics:")
    print(workload_df.describe())
    
    # Check for imbalance
    max_diff = workload_df["Total"].max() - workload_df["Total"].min()
    print(f"\nMaximum difference in total tasks: {max_diff}")
    
    return workload_df

# Analyze workload distribution
workload_df = analyze_workload(workload)

# Save workload analysis
workload_df.to_excel(f"{save_path}Donders_Wonders_Workload_Analysis.xlsx", index=False)

print("\nPlanning successfully generated and saved.")
```

```python
import pandas as pd
import numpy as np
import os
from collections import defaultdict
from datetime import timedelta, datetime

# Load data
dw_path = "/Users/wiegerscheurer/Library/CloudStorage/OneDrive-RadboudUniversiteit/Donders/Wonders/"
dw_files = os.listdir(dw_path)
team_filename = [f for f in dw_files if "team" in f][0]
planning_filename = [f for f in dw_files if "planning" in f][0]

team_raw_df = pd.read_excel(dw_path + team_filename)

# Function to extract first names if needed
def get_first_names(df):
    # This function should exist in your code
    # If it doesn't, replace with your actual logic for getting first names
    # For example:
    return df["Name"].apply(lambda x: x.split()[0] if isinstance(x, str) else x)

# Fix the team_df creation
team_df = team_raw_df.copy()
# Find the first null row to determine where data ends
if any(team_raw_df["Name"].isnull()):
    first_null_index = team_raw_df["Name"].isnull().idxmax()
    team_df = team_raw_df.loc[0:first_null_index-1].copy()
else:
    team_df = team_raw_df.copy()

team_df["Name"] = list(get_first_names(team_df))
# Sort the df based on the first name
team_df = team_df.sort_values(by="Name")

planning_df = pd.read_excel(dw_path + planning_filename)

# Define role weights
role_weights = {
    "Author": 5,
    "Buddy": 1,
    "Editor": 3,
    "Translator": 2,
    "Translation Editor": 1
}

# Generate dates
def generate_dates(start_date, end_date):
    start_date = datetime.strptime(start_date, "%Y-%m-%d")
    end_date = datetime.strptime(end_date, "%Y-%m-%d")
    num_weeks = (end_date - start_date).days // 7
    return [start_date + timedelta(weeks=i) for i in range(num_weeks + 1)]

# Calculate flexibility score (lower = less flexible)
def compute_flexibility(df, role_columns):
    df["Flexibility_Score"] = df[role_columns].apply(lambda row: sum(row == "yes"), axis=1)
    return df.sort_values(by="Flexibility_Score")

# Get all potential people for each role and compute flexibility for all groups
roles = ["Buddy", "Editor Dutch", "Editor English", "Translates English>Dutch", "Translates Dutch > English"]

# Identify Dutch and English authors
dutch_authors = team_df[(team_df["Blogs.1"] == "yes") & (team_df["Editor Dutch"] == "yes")].copy()
english_authors = team_df[(team_df["Blogs.1"] == "yes") & (team_df["Editor Dutch"] != "yes")].copy()
all_authors = team_df[team_df["Blogs.1"] == "yes"].copy()

# Compute flexibility scores
dutch_authors = compute_flexibility(dutch_authors, roles)
english_authors = compute_flexibility(english_authors, roles)
all_authors = compute_flexibility(all_authors, roles)

# Get people for each role with proper language skills
dutch_editors = team_df[team_df["Editor Dutch"] == "yes"].copy()
dutch_editors = compute_flexibility(dutch_editors, roles)

english_editors = team_df[team_df["Editor English"] == "yes"].copy()
english_editors = compute_flexibility(english_editors, roles)

translators_en_nl = team_df[team_df["Translates English>Dutch"] == "yes"].copy()
translators_en_nl = compute_flexibility(translators_en_nl, roles)

translators_nl_en = team_df[team_df["Translates Dutch > English"] == "yes"].copy()
translators_nl_en = compute_flexibility(translators_nl_en, roles)

dutch_buddies = team_df[(team_df["Buddy"] == "yes") & (team_df["Editor Dutch"] == "yes")].copy()
dutch_buddies = compute_flexibility(dutch_buddies, roles)

english_buddies = team_df[(team_df["Buddy"] == "yes") & (team_df["Editor English"] == "yes")].copy()
english_buddies = compute_flexibility(english_buddies, roles)

# Function to get the person with fewest weighted assignments
def get_next_person(eligible_people, workload, role):
    if eligible_people.empty:
        return None
        
    # Sort eligible people by weighted workload
    eligible_people_sorted = eligible_people.copy()
    
    # Add current role-specific workload to the dataframe
    eligible_people_sorted["Current_Role_Workload"] = eligible_people_sorted["Name"].apply(
        lambda name: workload.get(name, {}).get(role, 0)
    )
    
    # Calculate weighted total workload
    eligible_people_sorted["Weighted_Total_Workload"] = eligible_people_sorted["Name"].apply(
        lambda name: sum(count * role_weights.get(r, 1) for r, count in workload.get(name, {}).items())
    )
    
    # Sort by current role workload, then weighted total workload
    eligible_people_sorted = eligible_people_sorted.sort_values(
        by=["Current_Role_Workload", "Weighted_Total_Workload"]
    )
    
    # Return the name of the person with the lowest workload
    return eligible_people_sorted.iloc[0]["Name"]

# Function to check if Dutch authors are overloaded compared to English authors
def are_dutch_authors_overloaded(workload, dutch_author_names, english_author_names):
    # Calculate average author weighted workload for each group
    dutch_workload = [sum(count * role_weights.get(r, 1) for r, count in workload.get(name, {}).items()) 
                      for name in dutch_author_names if name in workload]
    
    english_workload = [sum(count * role_weights.get(r, 1) for r, count in workload.get(name, {}).items()) 
                        for name in english_author_names if name in workload]
    
    # Check if there are any Dutch authors
    if not dutch_workload:
        return False
    
    # Check if there are any English authors
    if not english_workload:
        return True
    
    # Calculate averages
    dutch_avg = sum(dutch_workload) / len(dutch_workload) if dutch_workload else 0
    english_avg = sum(english_workload) / len(english_workload) if english_workload else 0
    
    # If Dutch authors have 20% more workload than English authors, consider them overloaded
    threshold = 1.2
    
    # Return True if Dutch authors are overloaded
    return dutch_avg > (english_avg * threshold)

def assign_roles(start_date, end_date):
    dates = generate_dates(start_date, end_date)
    
    # Initialize workload tracking
    workload = defaultdict(lambda: defaultdict(int))
    
    # Initialize assignments dictionary
    assignments = defaultdict(list)
    
    # Get names for checking workload balance
    dutch_author_names = dutch_authors["Name"].tolist()
    english_author_names = english_authors["Name"].tolist()
    
    # Initialize language counters for tracking balance
    dutch_count = 0
    english_count = 0
    
    # For each date, assign all roles
    for date in dates:
        # Check if Dutch authors are overloaded
        dutch_overloaded = are_dutch_authors_overloaded(workload, dutch_author_names, english_author_names)
        
        # Determine the language for this publication
        # Try to balance languages while avoiding overloading Dutch authors
        if dutch_overloaded:
            # If Dutch authors are overloaded, use English author
            use_dutch = False
        elif english_count > dutch_count + 3:
            # If English is getting too dominant, prioritize Dutch
            use_dutch = True
        elif dutch_count > english_count + 1:
            # If Dutch is getting too dominant, prioritize English
            use_dutch = False
        else:
            # Otherwise, alternate with slight preference for balance
            use_dutch = english_count >= dutch_count
        
        # Select author pool based on language decision
        if use_dutch and not dutch_authors.empty:
            author_pool = dutch_authors.copy()
            is_dutch = True
        elif not english_authors.empty:
            author_pool = english_authors.copy()
            is_dutch = False
        else:
            # Fallback to any author
            author_pool = all_authors.copy()
            is_dutch = True  # Default assumption, will be checked after author selection
        
        # Get author for this date - prioritize those with fewer weighted assignments
        author = get_next_person(author_pool, workload, "Author")
        
        if author is None:
            print(f"No available author for date {date}")
            continue
        
        # Double-check language if we used the fallback
        if author_pool is all_authors:
            is_dutch = team_df[team_df["Name"] == author]["Editor Dutch"].values[0] == "yes"
            
        # Update workload for this author
        workload[author]["Author"] += 1
        
        # Update language counters
        if is_dutch:
            dutch_count += 1
        else:
            english_count += 1
        
        # Select appropriate role pools based on author's language
        if is_dutch:
            # Dutch author needs Dutch buddy, Dutch editor, Dutch->English translator
            buddy_pool = dutch_buddies.copy()
            editor_pool = dutch_editors.copy()
            translator_pool = translators_nl_en.copy()
            translation_editor_pool = english_editors.copy()
        else:
            # English author needs English buddy, English editor, English->Dutch translator
            buddy_pool = english_buddies.copy()
            editor_pool = english_editors.copy()
            translator_pool = translators_en_nl.copy()
            translation_editor_pool = dutch_editors.copy()
        
        # Exclude the author from other roles for this date
        buddy_pool = buddy_pool[buddy_pool["Name"] != author]
        editor_pool = editor_pool[editor_pool["Name"] != author]
        translator_pool = translator_pool[translator_pool["Name"] != author]
        translation_editor_pool = translation_editor_pool[translation_editor_pool["Name"] != author]
        
        # If any pool is empty, expand to include all qualified people
        if buddy_pool.empty:
            print(f"Warning: No appropriate buddy for {author}'s language. Using any available buddy.")
            buddy_pool = team_df[team_df["Buddy"] == "yes"].copy()
            buddy_pool = buddy_pool[buddy_pool["Name"] != author]
        
        if editor_pool.empty:
            print(f"Warning: No appropriate editor for {author}'s language. Using any available editor.")
            editor_pool = (dutch_editors.copy() if is_dutch else english_editors.copy())
            editor_pool = editor_pool[editor_pool["Name"] != author]
        
        if translator_pool.empty:
            print(f"Warning: No appropriate translator for {author}'s language. Using any available translator.")
            translator_pool = (translators_nl_en.copy() if is_dutch else translators_en_nl.copy())
            translator_pool = translator_pool[translator_pool["Name"] != author]
        
        if translation_editor_pool.empty:
            print(f"Warning: No appropriate translation editor for {author}'s language. Using any available editor.")
            translation_editor_pool = (english_editors.copy() if is_dutch else dutch_editors.copy())
            translation_editor_pool = translation_editor_pool[translation_editor_pool["Name"] != author]
        
        # Get people for each role based on weighted workload
        buddy = get_next_person(buddy_pool, workload, "Buddy")
        
        # Now exclude buddy from other roles
        editor_pool = editor_pool[editor_pool["Name"] != buddy]
        translator_pool = translator_pool[translator_pool["Name"] != buddy]
        translation_editor_pool = translation_editor_pool[translation_editor_pool["Name"] != buddy]
        
        editor = get_next_person(editor_pool, workload, "Editor")
        
        # Exclude editor from other roles
        translator_pool = translator_pool[translator_pool["Name"] != editor]
        translation_editor_pool = translation_editor_pool[translation_editor_pool["Name"] != editor]
        
        translator = get_next_person(translator_pool, workload, "Translator")
        
        # Exclude translator from translation editor role
        translation_editor_pool = translation_editor_pool[translation_editor_pool["Name"] != translator]
        
        translation_editor = get_next_person(translation_editor_pool, workload, "Translation Editor")
        
        # Update workloads
        workload[buddy]["Buddy"] += 1
        workload[editor]["Editor"] += 1
        workload[translator]["Translator"] += 1
        workload[translation_editor]["Translation Editor"] += 1
        
        # Calculate dates
        submit_to_buddy_date = date.strftime("%m/%d/%Y")
        submit_to_editor_translator_date = (date + timedelta(days=7)).strftime("%m/%d/%Y")
        return_to_author_date = (date + timedelta(days=19)).strftime("%m/%d/%Y")
        publish_date = (date + timedelta(days=24)).strftime("%m/%d/%Y")
        
        # Add to assignments
        assignments["Author"].append(author)
        assignments["Language"].append("Dutch" if is_dutch else "English")
        assignments["Submit to Buddy"].append(submit_to_buddy_date)
        assignments["Buddy"].append(buddy)
        assignments["Submit to Editor and Translator"].append(submit_to_editor_translator_date)
        assignments["Editor"].append(editor)
        assignments["Translator"].append(translator)
        assignments["Translate_into"].append("English" if is_dutch else "Dutch")
        assignments["Translation Editor"].append(translation_editor)
        assignments["Return to Author"].append(return_to_author_date)
        assignments["Publish"].append(publish_date)
    
    return assignments, workload, dutch_count, english_count

# Run the assignment algorithm
assignments, workload, dutch_count, english_count = assign_roles("2025-01-03", "2025-12-04")

# Create and save the assignments DataFrame
assignments_df = pd.DataFrame(assignments)
save_path = dw_path + "output/"
os.makedirs(save_path, exist_ok=True)
assignments_df.to_excel(f"{save_path}Donders_Wonders_Planning_Output.xlsx", index=False)

# Display workload distribution with weightings
def analyze_workload(workload, dutch_count, english_count):
    # Convert the nested defaultdict to DataFrame for analysis
    workload_rows = []
    for person, roles in workload.items():
        row = {"Name": person}
        raw_total = 0
        weighted_total = 0
        for role, count in roles.items():
            row[role] = count
            raw_total += count
            weighted_total += count * role_weights.get(role, 1)
        row["Raw_Total"] = raw_total
        row["Weighted_Total"] = weighted_total
        workload_rows.append(row)
    
    workload_df = pd.DataFrame(workload_rows)
    
    # Fill NaN with 0
    for col in workload_df.columns:
        if col != "Name":
            workload_df[col] = workload_df[col].fillna(0).astype(int)
    
    # Sort by weighted total workload
    workload_df = workload_df.sort_values(by="Weighted_Total")
    
    # Add language information
    workload_df["Is_Dutch_Author"] = workload_df["Name"].apply(
        lambda name: name in dutch_authors["Name"].values
    )
    
    # Get workload stats by language group
    dutch_authors_workload = workload_df[workload_df["Is_Dutch_Author"]]["Weighted_Total"]
    english_authors_workload = workload_df[~workload_df["Is_Dutch_Author"]]["Weighted_Total"]
    
    print("\n--- Workload Analysis ---")
    print("Role Weights:")
    for role, weight in role_weights.items():
        print(f"{role}: {weight}")
    
    print(f"\nLanguage Distribution:")
    print(f"Dutch articles: {dutch_count} ({dutch_count/(dutch_count+english_count)*100:.1f}%)")
    print(f"English articles: {english_count} ({english_count/(dutch_count+english_count)*100:.1f}%)")
    
    print("\nAssignments with Weighted Workload:")
    print(workload_df)
    
    print("\nWorkload Statistics:")
    print("Raw Total (unweighted):")
    print(workload_df["Raw_Total"].describe())
    print("\nWeighted Total:")
    print(workload_df["Weighted_Total"].describe())
    
    print("\nWorkload by Language Group:")
    if not dutch_authors_workload.empty:
        print(f"Dutch authors average weighted workload: {dutch_authors_workload.mean():.2f}")
    else:
        print("No Dutch authors in workload")
    
    if not english_authors_workload.empty:
        print(f"English authors average weighted workload: {english_authors_workload.mean():.2f}")
    else:
        print("No English authors in workload")
    
    # Check for imbalance
    max_raw_diff = workload_df["Raw_Total"].max() - workload_df["Raw_Total"].min()
    max_weighted_diff = workload_df["Weighted_Total"].max() - workload_df["Weighted_Total"].min()
    print(f"\nMaximum difference in raw tasks: {max_raw_diff}")
    print(f"Maximum difference in weighted tasks: {max_weighted_diff}")
    
    return workload_df

# Analyze workload distribution
workload_df = analyze_workload(workload, dutch_count, english_count)

# Save workload analysis
workload_df.to_excel(f"{save_path}Donders_Wonders_Workload_Analysis.xlsx", index=False)

print("\nPlanning successfully generated and saved.")
```
